```python
def death_of_json(row,col,col_country):
    if (sheet.isnull().iloc[row,col+1] or col == 4):
        if (sheet.iloc[row,col] == sheet.iloc[row + 1,col]):
            json_output.write('"' + sheet.iloc[row,col] + '":[{')
            json_output.write('"count": "')
            json_output.write(str(sheet.iloc[row,col_country]))
            json_output.write('",')
            death_of_json(row+1,col+1,col_country)
        elif (sheet.iloc[row,col] != sheet.iloc[row + 1,col]):
            json_output.write(',"' + str(sheet.iloc[row,col]) + '":')
            json_output.write('"' + str(sheet.iloc[row,col_country]) + '"')
            death_of_json(row+1,col,col_country)
        elif(sheet.isnull().iloc[row,col]):
            if(sheet.notnull().iloc[row,col-1]):
                json_output.write('}],')
                death_of_json(row+1,col-1,col_country)
            else :
                json_output.write('}]')
                death_of_json(row,col-1,col_country)
```
